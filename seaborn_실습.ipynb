```python
import pandas as pd
import seaborn as sns

targetUrl_drink="https://raw.github.com/vkdl/gitpro2/main/caffeine.csv"
drinkData = pd.read_csv(targetUrl_drink)

sns.set_theme()
sns.set(rc={"figure.dpi":300}) 
sns.set(rc={"figure.figsize":(6,3)}) 

sns.scatterplot(x="Volume (ml)", y ="Caffeine (mg)", data=drinkData, hue ='type')
```

```python
sns.histplot( y= "Calories", data = drinkData)
```

```python
sns.histplot( x= "Calories", data = drinkData)
```

```python
sns.histplot( x= "Calories", data = drinkData, kde = "True")
```

```python
sns.histplot( x= "Calories", data = drinkData, hue = "type")
```

```python
import matplotlib.pyplot as plt
sns.barplot(x="type", y="Calories", data =drinkData)
plt.xticks(rotation = 45)
```

```python
import matplotlib.pyplot as plt
plt.xticks(rotation = 45)
sns.boxplot(x="type", y="Caffeine (mg)", data=drinkData)
```
